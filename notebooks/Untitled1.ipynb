```python
import os
import json
import pandas as pd
from pathlib import Path
```

```python
DATA_FOLDER_PATH = Path("Olivetti_denoise_info_015_3000_PCA")
```

```python
all_data = list()
for rank in range(1, 201):
    file_name = f"UTK-PCA-3000-0_15-{rank}.json"
    file_path = os.path.join(DATA_FOLDER_PATH, file_name)
    try:
        with open(file_path, "r") as f:
            row_data = json.load(f)
        row_data["rank"] = rank
        all_data.append(row_data)
    except:
        pass
```

```python
df = pd.DataFrame(all_data)
df
```
