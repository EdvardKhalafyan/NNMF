```python
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
swimmer_data_path = "../data/results/appendix_swimmer_report.xlsx"
swimmer_noisy_data_path = "../data/results/appendix_swimmer_noisy_report.xlsx"
olivetti_data_path = "../data/results/appendix_olivetti_report.xlsx"
utk_data_path = "../data/results/appendix_utk_report.xlsx"

df_swimmer = pd.read_excel(swimmer_data_path)
df_swimmer_noisy = pd.read_excel(swimmer_noisy_data_path)
df_olivetti = pd.read_excel(olivetti_data_path)
df_utk = pd.read_excel(utk_data_path)
```

```python
labels = {
    "I_BI":          r"$I(\mathcal{B}:\mathcal{I})$",
    "I_BPi":         r"$I(\mathcal{B}:\Pi)$",
    "I_PiI":         r"$I(\Pi:\mathcal{I})$",
    "Ihat_PiI":      r"$\hat I(\Pi:\mathcal{I})$",
    "H_Pi_given_I":  r"$H(\Pi\,|\,\mathcal{I})$",
    "Hhat_Pi_given_I": r"$\hat H(\Pi\,|\,\mathcal{I})$",
    "H_B":           r"$H(\mathcal{B})$",
    "H_I":           r"$H(\mathcal{I})$",
    "H_Pi":          r"$H(\Pi)$",
    "Ntilde_BI":     r"$\tilde I(\mathcal{B}:\mathcal{I})$",
    "Ntilde_BPi":    r"$\tilde I(\mathcal{B}:\Pi)$",
    "Ncirc_BI":      r"$I^{\circ}(\mathcal{B}:\mathcal{I})$",
    "Ncirc_BPi":     r"$I^{\circ}(\mathcal{B}:\Pi)$",
}
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=False)

# --- Swimmer ---
ax = axes[0, 0]
ax.plot(df_swimmer["rank"], df_swimmer["I_BI"], label=labels["I_BI"])
ax.plot(df_swimmer["rank"], df_swimmer["I_BPi"], label=labels["I_BPi"])
ax.plot(df_swimmer["rank"], df_swimmer["I_PiI"], label=labels["I_PiI"])
ax.set_title("Swimmer")
ax.set_xlabel("rank")
ax.legend()
ax.grid(True)

# --- Swimmer noisy ---
ax = axes[0, 1]
ax.plot(df_swimmer_noisy["rank"], df_swimmer_noisy["I_BI"], label=labels["I_BI"])
ax.plot(df_swimmer_noisy["rank"], df_swimmer_noisy["I_BPi"], label=labels["I_BPi"])
ax.plot(df_swimmer_noisy["rank"], df_swimmer_noisy["I_PiI"], label=labels["I_PiI"])
ax.set_title("Swimmer (noise=0.25)")
ax.set_xlabel("rank")
ax.legend()
ax.grid(True)

# --- Olivetti ---
ax = axes[1, 0]
ax.plot(df_olivetti["rank"], df_olivetti["I_BI"], label=labels["I_BI"])
ax.plot(df_olivetti["rank"], df_olivetti["I_BPi"], label=labels["I_BPi"])
ax.plot(df_olivetti["rank"], df_olivetti["I_PiI"], label=labels["I_PiI"])
ax.set_title("Olivetti")
ax.set_xlabel("rank")
ax.legend()
ax.grid(True)

# --- UTK ---
ax = axes[1, 1]
ax.plot(df_utk["rank"], df_utk["I_BI"], label=labels["I_BI"])
ax.plot(df_utk["rank"], df_utk["I_BPi"], label=labels["I_BPi"])
ax.plot(df_utk["rank"], df_utk["I_PiI"], label=labels["I_PiI"])
ax.set_title("UTK")
ax.set_xlabel("rank")
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.savefig("combined_2x2_plots.pdf")
plt.show()
```
