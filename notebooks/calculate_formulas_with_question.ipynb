```python
import os
import data
import random
import sklearn
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.utils import shuffle, check_random_state

import config
from utils import apply_symmetric_noise

random.seed(config.SEED)
np.random.seed(config.SEED)
np.random.default_rng(config.SEED)
check_random_state(config.SEED)

import warnings

warnings.filterwarnings("ignore")
```

```python
array_swimmer = shuffle(data.load_swimmer())
array_olivetti = shuffle(data.load_olivetti())
array_utk = shuffle(data.load_utk())

array_swimmer_noisy = apply_symmetric_noise(array_swimmer, prob=0.25)
```

```python
swimmer_noisy_ranks = list(range(8, 184, 4))
swimmer_ranks = list(range(2, 35))
olivetti_ranks = [36, 54, 72, 90, 108, 126, 144, 162, 180]
# olivetti_ranks = list(range(2, 100, 15))
utk_ranks = [36, 72, 108]
```

```python
swimmer_results = list()
for rank in tqdm(swimmer_ranks):
    res = check_appendix_relations(
        P_images_by_pixels=array_swimmer,
        rank=rank,
        seed=config.SEED,
        loss="frob",
        base=np.e,
        tol=1e-9,
        transpose_input=True,
        report=False
    )
    res["rank"] = rank
    swimmer_results.append(res)

df_swimmer = pd.DataFrame(swimmer_results)
df_swimmer.insert(0, "rank", df_swimmer.pop("rank"))
df_swimmer.to_excel("appendix_swimmer_report.xlsx", index=False)
```

```python
swimmer_noisy_results = list()
for rank in tqdm(swimmer_noisy_ranks):
    res = check_appendix_relations(
        P_images_by_pixels=array_swimmer_noisy,
        rank=rank,
        seed=config.SEED,
        loss="frob",
        base=np.e,
        tol=1e-9,
        transpose_input=True,
        report=False
    )
    res["rank"] = rank
    swimmer_noisy_results.append(res)

df_swimmer_noisy = pd.DataFrame(swimmer_noisy_results)
df_swimmer_noisy.insert(0, "rank", df_swimmer_noisy.pop("rank"))
df_swimmer_noisy.to_excel("appendix_swimmer_noisy_report.xlsx", index=False)
```
